```python
from dual import *
import sympy
from functions import *
import math
from collections.abc import Iterable
from tuples import *
from functools import reduce
from typing import Callable
from derivatives import *
```

```python
def Gamma(q: Callable) -> Callable:
    def Gq(*args):
        return up(
            *args,
            q(*args),
            D(q)(*args)
        )
    return Gq
```

```python
def q(t):
    return up(t, t**2, sin(t**3))
```
